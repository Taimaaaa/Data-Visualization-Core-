## Explore BMI

### 9\) Regplot - Charges vs. BMI

9a) Make a seaborn regplot to visualize the relationship between charges and bmi for the entire dataset

```python
# Make a seaborn regplot for bmi vs. charges.
scatter_kws = dict(edgecolor="white")
line_kws= dict(color="black")
ax = sns.regplot(data=df, x="bmi", y="charges", scatter_kws=scatter_kws, line_kws=line_kws)
ax.set_title("BMI vs Charges Regplot");
```

9b\) What do you notice from the visualization? *(There are many correct answers.)*

- Your Answer:
  - `There’s a weak positive trend: as BMI increases, charges tend to rise, but the scatter is wide, so BMI alone explains little. The cloud fans out: low BMI has mostly <$15k, while higher BMI shows a much bigger spread, including many points $30k–$50k+ and a few >$60k around BMI ~30–35. The plot also hints at two bands (low vs high charges), suggesting another variable (such as smoker) is driving the split.`

### 11\)  Linear Model Plot - Charges vs. BMI - Colored by Smoker



11a) Compare the relationship of bmi and charges for smokers and nonsmokers using seaborn's `lmplot` to calculate a separate trendline for Smokers vs. Non-Smokers.


```python
# Create an lmplot of bmi vs. charges, colored by smoker.
g = sns.lmplot(data=df, x="bmi", y="charges", hue="smoker", scatter_kws=dict(edgecolor="white"))
g.ax.set_title("BMI vs Charges colored by smoker LMPlot");
```

11b) What do you notice from the visualization? (Compare the trendlines.)

- Your Answer:
  - `The trendlines split clearly: for smokers (orange) the line is much higher and steeply positive; charges rise sharply as BMI increases. while for non-smokers (blue) the line is low and almost flat, meaning BMI has only a small effect on charges. At any given BMI, smokers’ expected charges are far above non-smokers’. The different slopes suggest an interaction (BMI matters mainly among smokers).`

14a) Calculate the ratio of smokers in each region and save it as a variable called "percent_smoker_by_region". (Hint: use .groupby with "region" combined with slicing the "smoker" column and end with .mean() )
